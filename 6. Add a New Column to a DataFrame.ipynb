# 1.2 Adding a new column with an array

### The dataframe has 1000 rows, so we need to create an array with 1000 elements

```python
# import numpy
import numpy as np
```

```python
# create an array of 1000 elements
language_score = np.arange(0, 1000)
```

```python
# length of the column
len(language_score)
```

```python
# adding a new column to the dataframe with an array
df_exams["language score"] = language_score 
df_exams
```

```python
# create random integer numbers between 1 and 100
int_language_score = np.random.randint(1, 100, size = 1000)
```

```python
# min value inclusive and high value exclusive
print(min(int_language_score))
print(max(int_language_score))
```

```python
# Adding a new column to the dataframe of an array
df_exams["language score"] = int_language_score 

pd.set_option("display.max_rows", 1000)

df_exams
```
